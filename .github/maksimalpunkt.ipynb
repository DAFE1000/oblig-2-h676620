```python
import math
import matplotlib.pyplot as plt

#Definerer funksjonen
def f(x):
    return math.exp(-x/4) * math.atan(x)
```

```python
#Ligningen som bestemmer toppunktet:
def g(x):
    return math.atan(x) - 4/(x**2 + 1)
```

For newtons metode trenger vi den deriverte av g som er:
g'(x) = 1 / (1 + x^2) + 8x / (x^2 + 1)^2

```python
#den deriverte av g
def g_deriv(x):
    return 1/(1 + x**2) + 8*x/(x**2 + 1)**2
```

```python
#Newtons metode

x = 1.0 #startverdi

for i in range(20):
    x = x - g(x)/g_deriv(x)

# x-kordinaten til toppunktet
x_topp = x

# y-kordinaten til toppunktet
y_topp = f(x_topp)

print(f"x-kordinat: {x_topp:.6f}")
print(f"y-kordinat: {y_topp:.6f}")
```

```python
#Plotting:

#lager x-verdier og y-verdier til plottet
x_verdier = [i/100 for i in range(-200, 801)]
y_verdier = [f(x) for x in x_verdier]

#plotter funkjsonen
plt.plot(x_verdier, y_verdier, label="f(x)")

#markerer toppunktet
plt.plot(x_topp, y_topp, 'o', label=f"Toppunkt ({x_topp:.4f}, {y_topp:.4f})")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("graf av f(x) med toppunkt")
plt.grid(True)
plt.legend()
plt.show()
```

# Svar

Ved numerisk løsning av ligningen
 $$
 \tan^{-1}(x) - \frac{4}{x^2+1} = 0
 $$

 fant vi toppunktet:

 $$
 (x, f(x)) \approx (1.6907,\; 0.6793)
 $$
